```python
# exercise 1

prime_number_list = [1,3,5,7,9,11,13,17,19,23]

# adding a item
prime_number_list.append(29)
print(prime_number_list)
```

```python
# deleting a item
del prime_number_list[4]
print(prime_number_list)
```

```python
# largest prime in the given list
largest_prime = max(prime_number_list)
# smallest prime in the given list
smallest_prime = min(prime_number_list)

print(largest_prime)
print (smallest_prime)
```

```python
# exercise 2(creating and reversing a tuple)

my_friends = ("vignesh","sankar","sreerag")
print (my_friends)
```

```python
# reversing the tuple

def Reverse(given_tuples): 
    new_tup = given_tuples[::-1] 
    return new_tup 
print(Reverse (my_friends))
```
